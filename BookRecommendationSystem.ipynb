## Popularity Model

```python
book_userrating = pd.merge(books, ratings, on='ISBN')
book_userrating.rename(columns={'Book-Title':'book_title','Book-Rating':'bookrating'}, inplace=True)
book_userrating.head()
```

```python
avg_rating=book_userrating.groupby('book_title').bookrating.mean().reset_index()
avg_rating.head()
```

```python
avg_rating=book_userrating.groupby('book_title').bookrating.agg(['count', 'mean']).reset_index()
avg_rating.rename(columns={'book_title':'Book-Title'}, inplace=True)
avg_rating.head()
```

```python
avg_rating.rename(columns={'count':'num_rating','mean':'avg_rating'}, inplace=True)
book_rating = pd.merge(books, avg_rating, on='Book-Title')
book_rating.head()
```

```python
rel_books=book_rating[book_rating['num_rating'] >= 100]
rel_books.head()
```

```python
popularbooks=rel_books.sort_values('avg_rating',ascending=False).head(200)
popularbooks
```

```python
popularbooks=popularbooks.drop_duplicates(subset=['Book-Title'])
popularbooks
```

## Collobrative Filtering Method

```python
x = ratings['User-ID'].value_counts() > 200
y=x[x].index
rel_user_book=book_userrating[book_userrating['User-ID'].isin(y)]
rel_user_book
```

```python
z = rel_user_book.groupby('book_title').count()['bookrating']>=50
a=z[z].index
rel_user_book=rel_user_book[rel_user_book['book_title'].isin(a)]
rel_user_book
```

```python
rqd_table = rel_user_book.pivot_table(index='book_title',columns='User-ID',values='bookrating')
rqd_table=rqd_table.fillna(0)
rqd_table
```

```python
from sklearn.neighbors import NearestNeighbors

model=NearestNeighbors()
model.fit(rqd_table)
```

```python
def get_recommend(bookname):
    giv_index = np.where(rqd_table.index==bookname)[0][0]
    distances, suggestions = model.kneighbors(rqd_table.iloc[giv_index, :].values.reshape(1, -1))
    for i in range(len(suggestions)):
        print(rqd_table.index[suggestions[i]])
```

```python
get_recommend("The Da Vinci Code")
```
